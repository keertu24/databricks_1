```python
corrupt_schema_1="emp_id INTEGER, Name STRING, Age INTEGER, Salary DOUBLE ,Department STRING,_corrupt_record STRING"
```

```python
df=spark.read.format("csv").\
        option("header","true").\
            schema(corrupt_schema_1) .\
                option("mode","permissive").\
                option("columnNameOfCorruptRecord", "_corrupt_record").\
                    load("/Volumes/youtube/raja_de/files/Corrupt.csv")
display(df)
```
