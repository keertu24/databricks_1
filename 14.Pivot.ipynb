```python

data = [
    ("India",  "Electronics", 1000),
    ("India",  "Clothing",    500),
    ("India",  "Grocery",     200),
    ("USA",    "Electronics", 1500),
    ("USA",    "Clothing",    700),
    ("USA",    "Grocery",     400),
    ("UK",     "Electronics", 900),
    ("UK",     "Clothing",    300),
    ("UK",     "Grocery",     250)
]

columns = ["country", "category", "sales"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
pivot_df=df.groupBy("country").pivot("category").sum("sales")
```

```python
display(pivot_df)
```

```python
unpivot_df=pivot_df.unpivot(
    ids=["country"],
   values=["Electronics", "Clothing", "Grocery"],
    variableColumnName="category",
    valueColumnName="sales"
)
display(unpivot_df)
```
