```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ColabSpark") \
    .getOrCreate()
```

```python
df = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/Pyspark_file/Baby_Names.csv",header=True,inferSchema=True)
df.show()
```

```python
df.rdd.getNumPartitions()
```

```python
from pyspark.sql.functions import spark_partition_id
df1=df.repartition(8).withColumn("partitionId",spark_partition_id()).groupBy("partitionId").count()
df1.show()
```

```python
df2=df.repartition(20).withColumn("partitionId",spark_partition_id()).groupBy("partitionId").count()
df2.show()
```
