```python
simpleData = [
    ("James",   "Sales",      3000),
    ("Michael", "Sales",      4600),
    ("Robert",  "Sales",      4100),
    ("James",   "Sales",      3000),
    ("Saif",    "Sales",      4100),
    
    ("Maria",   "Finance",    3000),
    ("Scott",   "Finance",    3300),
    ("Jen",     "Finance",    3900),
    ("Alex",    "Finance",    3500),
    ("Nina",    "Finance",    4200),

    ("Jeff",    "Marketing",  3000),
    ("Kumar",   "Marketing",  2000),
    ("John",    "Marketing",  4500),
    ("Sara",    "Marketing",  3800),

    ("David",   "HR",         2800),
    ("Rita",    "HR",         3100),
    ("Sam",     "HR",         2900),
    ("Priya",   "HR",         3500)
]

columns = ["employee_name", "department", "salary"]

df = spark.createDataFrame(data=simpleData, schema=columns)
df.show()
```

```python
from pyspark.sql import Window
from pyspark.sql.functions import lead,desc,lag
windowRange=Window.partitionBy(df.department).orderBy("salary")
df1=df.withColumn("Lead Salary",lead(df.salary,2).over(windowRange))\
    .withColumn("Lag Salary",lag(df.salary).over(windowRange))
display(df1)
```
