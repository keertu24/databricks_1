```python
data_a = [
    (1, "Alice", 50000),
    (2, "Bob", 70000),
    (3, "Cara", 60000)
]
columns_a = ["id", "name", "salary"]

df_a = spark.createDataFrame(data_a, columns_a)
```

```python
data_b = [
    (3, "Cara", 60000),   # duplicate of df_a row
    (4, "David", 80000),
    (5, "Eve", 55000)
]
columns_b = ["id", "name", "salary"]

df_b = spark.createDataFrame(data_b, columns_b)
```

```python
data_c = [
    (1, "Alice", 50000),
    (2, "Bob", 70000),
    (3, "Cara", 60000)     # same id as in df_a but different structure
]
columns_c = ["id", "full_name", "department"]

df_c = spark.createDataFrame(data_c, columns_c)
```

```python
data_d = [
    (1, "Alice", 50000,"df"),
    (2, "Bob", 70000,"df"),
    (3, "Cara", 60000,"df")
]
columns_d = ["id", "name", "salary","col"]

df_d = spark.createDataFrame(data_d, columns_d)
```

```python

union_df=df_a.union(df_b)
display(union_df)
```

```python
union_all_df=df_a.unionAll(df_b)
display(union_all_df)
```

```python
union_df2=df_a.union(df_c)
display(union_df2)
```

```python
union_df3=df_c.union(df_a)
display(union_df3)
```

```python
df_union = df_a.unionByName(df_c, allowMissingColumns=True)
df_union.show()
```

```python
union_df5=df_a.unionByName(df_d, allowMissingColumns=True)
display(union_df5)
```
