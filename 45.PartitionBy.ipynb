```python
df=spark.read.format("csv").option("inferSchema",True).option("header",True).load("/Volumes/youtube/raja_de/partitions_example/baby_names_input/baby_names_20000.csv")
display(df)
```

##Distinct Year and Count for each year

```python
df.groupBy("Year").count().show(truncate=False)
```

##Partition By One Key Column

```python
df.write.option("header",True)\
    .partitionBy("Year")\
    .mode("overwrite")\
    .csv("/Volumes/youtube/raja_de/partitions_example/baby_names_output/")
```

##Partition By Multi Key Column

```python
df.write.option("header",True)\
    .partitionBy("Year","Sex")\
    .mode("overwrite")\
    .csv("/Volumes/youtube/raja_de/partitions_example/baby_names_output_multi/")
```

##Partition by Key Column along with maximum number of records for each partition 

```python
dbutils.fs.mkdirs("/Volumes/youtube/raja_de/partitions_example/baby_names_output_max_per_parti/")
```

```python
df.write.option("header",True)\
    .option("maxRecordPerFile",400)\
    .partitionBy("Year")\
    .mode("overwrite")\
    .csv("/Volumes/youtube/raja_de/partitions_example/baby_names_output_max_per_parti/")
```
