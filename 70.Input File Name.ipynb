```python
df = spark.read.option("header", True) \
        .csv("/Volumes/youtube/raja_de/input_file_name_example/yearly_quarters/*/*/*.csv")
display(df)
```

```python
display(df.filter(df.id.isNull()))
```

```python
from pyspark.sql.functions import input_file_name
df1=df.withColumn("File Name",input_file_name())
display(df1)
```

```python
from pyspark.sql.functions import col

df2 = df.select(
    "*",
    col("_metadata.file_path").alias("Source_File")
)

display(df2)
```

```python
display(df2.filter(df.id.isNull()))
```
