```python
empDF = [
    (
        'Sales_dept',
        [
            {'emp_name': 'John',  'salary': 1000, 'yrs_of_service': 10, 'Age': 33},
            {'emp_name': 'David', 'salary': 2000, 'yrs_of_service': 15, 'Age': 40},
            {'emp_name': 'Nancy', 'salary': 3000, 'yrs_of_service': 20, 'Age': 45},
            {'emp_name': 'Mike',  'salary': 3000, 'yrs_of_service': 6,  'Age': 30},
            {'emp_name': 'Rosy',  'salary': 6000, 'yrs_of_service': 8,  'Age': 32}
        ]
    ),
    (
        'HR_dept',
        [
            {'emp_name': 'Edwin', 'salary': 6000,  'yrs_of_service': 8,  'Age': 31},
            {'emp_name': 'Tomas', 'salary': 3000,  'yrs_of_service': 4,  'Age': 26},
            {'emp_name': 'Sarah', 'salary': 12000, 'yrs_of_service': 22, 'Age': 49},
            {'emp_name': 'Stella','salary': 15000, 'yrs_of_service': 25, 'Age': 52},
            {'emp_name': 'Kevin', 'salary': 4000,  'yrs_of_service': 5,  'Age': 27}
        ]
    )
]
```

```python
df_brand = spark.createDataFrame(
    data=empDF,
    schema=['Department', 'Employee']
)

df_brand.printSchema()
display(df_brand)
```

```python
from pyspark.sql.functions import arrays_zip
df_brandZip=df_brand.withColumn("Zip",arrays_zip(df_brand["Employee"]))
display(df_brandZip)
```

```python

from pyspark.sql.functions import explode, col

df_flat = df_brand.withColumn("emp", explode("Employee")) \
                  .select(
                      col("Department"),
                      col("emp.emp_name").alias("emp_name"),
                      col("emp.salary").alias("salary"),
                      col("emp.yrs_of_service").alias("yrs_of_service"),
                      col("emp.Age").alias("Age")
                  )

display(df_flat)
df_flat.printSchema()
```
