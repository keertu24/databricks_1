```python
input_data = [
    ("David",   70, 68, 89, 40, 84),
    ("Kevin",   90, 67, 87, 79, 74),
    ("Natalia", 66, 88, 49, 65, 72),
    ("Roger",   78, 73, 82, 89, 67),
    ("Michael", 80, 86, 69, 78, 92)
]

schema = ["Student", "Subject_1", "Subject_2", "Subject_3", "Subject_4", "Subject_5"]

df = spark.createDataFrame(input_data, schema)
df.display()
```

##Greast Columns

```python
from pyspark.sql.functions import greatest

greatDF=df.withColumn("Greastest Value",greatest( "Subject_1", "Subject_2", "Subject_3", "Subject_4", "Subject_5"))
display(greatDF)
```

##Least of Columns

```python
from pyspark.sql.functions import least

leastDF=df.withColumn("least Value",least( "Subject_1", "Subject_2", "Subject_3", "Subject_4", "Subject_5"))
display(leastDF)
```

##Max of Rows

```python
df.agg({"Subject_1":'max'}).show()
```

```python
df.agg({"Subject_1":'min',"Subject_2":'min',"Subject_3":'min',"Subject_4":'min',"Subject_5":'min'}).show()
```
