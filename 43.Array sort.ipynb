```python
data = [
    # John
    ("John", 9),
    ("John", 2),
    ("John", 7),
    ("John", 4),
    ("John", 6),

    # David
    ("David", 1),
    ("David", 7),
    ("David", 3),
    ("David", 5),
    ("David", 4),

    # Mike
    ("Mike", 6),
    ("Mike", 3),
    ("Mike", 9),
    ("Mike", 1),
    ("Mike", 2),

    # Sarah
    ("Sarah", 8),
    ("Sarah", 3),
    ("Sarah", 7),
    ("Sarah", 1),

    # Kevin
    ("Kevin", 5),
    ("Kevin", 7),
    ("Kevin", 2),

    # Emma
    ("Emma", 4),
    ("Emma", 6),
    ("Emma", 1)
]

schema = ["Name", "Value"]
df = spark.createDataFrame(data, schema)
display(df)
```

```python
from pyspark.sql.functions import collect_list

df_arr = df.groupBy("Name").agg(
    collect_list("Value").alias("Array_Values")
)
display(df_arr)
```

```python
from pyspark.sql.functions import array_sort

df_sorted = df_arr.withColumn(
    "Sorted_Array",
    array_sort("Array_Values")
)

display(df_sorted)
```

```python
from pyspark.sql.functions import array_sort, when, length, col

df = spark.createDataFrame([ (["foo","foobar","bar","a",None],) ], ["data"])

df_sorted = df.select(
    array_sort(
        col("data"),
        # comparator: compare by length descending
        lambda x, y: when(x.isNull() | y.isNull(), 0) \
                     .otherwise(length(y) - length(x))
    ).alias("sorted")
)

df_sorted.show(truncate=False)
```
