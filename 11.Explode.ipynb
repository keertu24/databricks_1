Creating Data Frame

```python
data_array = [
    (1, ["red", "blue", "green",None]),
    (2, [None]),
    (3, None),
    (4, ["yellow", None]),
]

df=spark.createDataFrame(data=data_array, schema=["id", "colors"])
display(df)
```

Explode Array

```python
from pyspark.sql.functions import explode

df2=df.select(df.id,explode(df.colors))
display(df2)

df.printSchema()
df2.printSchema()
```

Explode Outer

```python
from pyspark.sql.functions import explode_outer

df3=df.select(df.id,explode_outer(df.colors))
display(df3)

df.printSchema()
df3.printSchema()
```

Explode Map(key value)

```python
data_map = [
    (1, {"A": 10, "B": 20}),
    (2, {}),
    (3, None),
    (4, {"X": None, "Y": 40})
]

df_map=spark.createDataFrame(data=data_map, schema=["id", "data"])
display(df_map)
```

```python
df_map2=df_map.select(df_map.id,explode(df_map.data))
df_map2.printSchema()
display(df_map2)
```
