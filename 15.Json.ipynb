```python
raw_df=spark.read.option("multiline","true").json("/Volumes/youtube/raja_de/files/sample_data.json")
```

```python
display(raw_df)
```

```python
raw_df.printSchema()
```

```python
from pyspark.sql.functions import explode, col

df_emp = raw_df.select(explode(col("Employee")).alias("emp"))

df_emp.printSchema()
display(df_emp)
```

```python
df_emp_flat = df_emp.select(
    col("emp.emp_id"),
    col("emp.Designation"),
    explode(col("emp.attribute")).alias("attr")
)
df_emp_flat.printSchema()
display(df_emp_flat)
```

```python
df_final = df_emp_flat.select(
    "emp_id",
    "Designation",
    "attr.Parent_id",
    "attr.status_flag",
    explode(col("attr.Department")).alias("dept")
)

display(df_final)
df_final.printSchema()
```

```python
df_final = df_emp_flat.select(
    "emp_id",
    "Designation",
    "attr.Parent_id",
    "attr.status_flag",
    explode(col("attr.Department")).alias("dept")
).select(
    "emp_id",
    "Designation",
    "Parent_id",
    "status_flag",
    col("dept.Dept_id").alias("Dept_id"),
    col("dept.Code").alias("Code"),
    col("dept.dept_type").alias("dept_type"),
    col("dept.dept_flag").alias("dept_flag")
)

df_final.show(truncate=False)
df_final.printSchema()
```
