```python
data = [
    (1, "Alice", 45000, "HR", 3),
    (2, "Bob", 80000, "IT", 6),
    (3, "Charlie", 30000, "Finance", 2),
    (4, "David", 120000, "IT", 9),
    (5, "Eve", None, "HR", 4),
    (6, "Frank", 70000, None, 5),
    (7, "Grace", 25000, "Finance", 1),
]

columns = ["id", "name", "salary", "department", "experience"]

df=spark.createDataFrame(data, columns)
```

```python
from pyspark.sql.functions import when

df1=df.withColumn("Status",when(df.salary<40000,"Low").
                  when(df.salary<80000,"Mid").
                  otherwise("high")

                  )
display(df1)
                            
```
